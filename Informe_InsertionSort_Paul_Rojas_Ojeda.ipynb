# 1. Problema de ordenamiento 
**Entrada:** Tendremos un arreglo de n numeros (sin ningun orden en especifico) 

**Salida:** Obtendremos un secuencia ordenada de **menor a mayor** de tal maquera que se dara una secuencia 



**Insertion Sort** es un algoritmo eficiente para ordenar una cantidad de n elementos, este algoritmo es la forma que muchas personas ordenan sus cartas, este es ordenando de teniendo en la mano izquierda la primera carta y el resto en la mesa recogiendo de una a una las cartas y ordenandolas en la mano izquierda de menor a mayor.

# Insertion Sort
El siguiente código muestra una implementacion del algoritmo **Insertion Sort**

## 2.1 Código en Python ###

```python
import random
from termcolor import colored
import copy

def Insertion_Sort(arg, verbose=False):
  largo_ar = len(arg) 
  contador = 0 #Contador der comparaciones
  if verbose == True: 
    print("input array:", arg)
  
  for i  in range(1, largo_ar):

    
        # Flag que indica si el arreglo ya se encuentra ordenado
    already_sorted = True


    key = arg[i] #obtenemos el valor del indice siguiente indice 
    
    j = i - 1 #obtenemos el indice de partida 

    if verbose == True: 
      print("\nPasada",i)
      print(colored(f"{str(arg[:j+1])[1:-1]}","red"),",",colored(f"{str(arg[i:i+1])[1:-1]}","blue"),",",str(arg[i+1:])[1:-1])

    

    # Se recorre el arreglo
        #Buscamos que el indice no sea negativo
        # y el valor siguiente sea mayor 
    while j >= 0 and key < arg[j]: 


             
      contador +=1
      arg[j+1] = arg[j]
      j -= 1
      if verbose == True: 
        print("comparacion: ", contador)
        print(colored(f"{str(arg[:j+2])[1:-1]}","red"),",",colored(key,"blue"))


    arg[j+1] = key
    if verbose == True: 
      print("al finalziar la pasada: ")
      print(colored(f"{str(arg[:i])[1:-1]}","red"),",",colored(f"{str(arg[i:i+1])[1:-1]}","red"),",",str(arg[i+1:])[1:-1])
  if verbose == True: 
    print("\n is sorted!")
    print("\n output array:",arg)
  
  return arg, contador


# Ejemplos
arg1 = [5,2,43,51,3,1]
print("Entrada: ", arg1)
arg1,contador = Insertion_Sort(arg1)
print("Salida: ", arg1)
print("# comparaciones: ", contador)
```

### 2.2 Descripción del algoritmo

El algoritmo recibe como entrada un arreglo *arg* con la secuencia de *n* elementos que se busca ordenar.

Para ser ordenados dentro del mismo arreglo:

1.   Se busca un *pivote (arg[j])* que sera el valor que se considera el menor de el arreglo en ese momento, para compararlo con el siguiente elemento *key* que sera el valor a comparar si este es menor al *pivote* entonce este toma su posicion y de este modo se busca el siguiente pivote *j-=1* para demostrar que el *key* sea menor o mayor.
2. Despues de haber colacado los elementos de menor a mayor, se busca el nuevo pivote quien seria el mayor de los elementos ordenados, para comparar con el siguiente elemento del arreglo y ordenarlos siguiendo el paso 1.
3. Al final, se retorna el arreglo ordenado y un *contador* de comparaciones

Note que no es necesario recorrer todo el arreglo, basta comparar el nuevo elemento con el pivote. es decir que cada iteracion se tiene al momento de realizar un cambio.

# 3. Tiempo de ejeción
**Teorema (Tiempo de ejecución)**

El algoritmo *insertionSort* tiene tiempo de ejecución es $O(n^2)$ en el peor caso.

# Prueba del teorema
Observando el codigo, se puede apreciar que el *peor caso* ocurre cuando el arreglo esta ordenado de *mayor a menor*.
En este caso, en cada pasada el eelmento se moveria una casilla a la izquierda, por lo que se necesitaria realizar $n^2$ pasadas para que los elementos quedaran ordenados. 

### Ejemplo

```python
a =[5,4,3,2];
print("Entrada no ordenada:",a)
a,counter=Insertion_Sort(a,True)
print("Salida ordenada:",a)
print("Total de comparaciones realizadas:",counter)
```

En este caso, la primera pasada hizo 1 comparacion, la segunda 2, la tercera 3. Total = 6

En el caso general, se deberían realizar 


$
\sum_{i=0}^{n-1} i= \frac{n(n-1)}{2}\qquad  
$ comparaciones. Por lo que el **Tiempo de ejecución del algoritmo en el peor caso es $O(n^2)$**



# 4. Correctitud
**Teorema (Correctitud)**
El algoritmo *InsertionSort* genera un arreglo de $[x_1,x_2,...,x_n]$, con los mismo elementos del arreglo de entradas pero esto ordenados de menor a mayor 
$x_1 <= x_2 <= ... <= x_n $

## Prueba del Teorema

Observamos que en cada iteracion de *insertSort* realizamos un cambio de orden entre el elemento selecionado y el pivote de los elementos ordenados.

Por lo que se probara el siguiente **propiedad invariante de bucle** para *insertion sort:*


*  Al comenzar cada iteracion j del bucle más grande, los ultimos j-1 elementos del arreglo corresponde a los mayores elementos del arreglo original ordenados de menor a mayor.


## Inicialización
Al comienzo de la primera iteracion(i=1), los ultimos *j= i-1 = 0* elementos cumplen la propiedad.

## Mantención

Aplicando inducción, asumimos que al comienzo de cada iteración *i* se cumple la propiedad, es decir los ultimos elementos *j - 1* del arreglo que corresponde a los menores elementos del arreglo en ese momento, de este modo en cada iteracion se ordenaran de menor a mayor.

Cosideramos que "X" como nuestro numero menor en ese momento y lo comparamos con el siguiente valor de "Y" de este modo se ordena, entre ellos quedado en nuestra izquierda los elementos ordenados de menor a mayor en ese momento para pasar.

Al finalizar la iteración (y al comienzo de la siguiente), se cunplira la propiedad invariante.


**Correctitud**

Finalmente, como la propiedad de bucle invariante es verdadera al incio del bucle y se mantiene en cada iteracion, se dice que al finalizar la n-ésima iteración del areglo, se genera un arreglo con los mismo elementos de entradas odenados de menor a mayor.
