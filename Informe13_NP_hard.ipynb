## ¿Qué quiere decir que un problema es NP-Hard? 

Los problemas **NP-hard**, son problemas con una complejidad comparable a los **NP-completos**, pero que no necesariamente pertenecen al conjunto **NP**.
Un problema $B$ es NP-hard, si cualquier problema $A$ de naturaleza NP-hard es reducible a $B$ en tiempo polinomial.

Los problemas NP-hard no están limitados a los problemas de decisión, sino que también incluyen problemas de optimización o de cualquier otro tipo. Si un problema $A$ es NP-hard, cualquier problema en el conjunto NP se puede reducir en tiempo polinomial al problema $A$.

Se dice que los problemas NP-hard son como minimo tan dificiles como los problemas más difíciles del conjunto NP (pero no necesariamente en NP).











## ¿Qué quiere decir que $P \neq NP$ ?

Primero entenderemos el significado de los conjuntos $P$ y $NP$. 

Para ello, clasificaremos los tipos de problemas computacionales de acuerdo a la forma de resolverlos:

<br>

<br>

Un problema **decidible** es aquel que posee un *algoritmo* que puede resolverlo. Estos problemas se subdividen en problemas **tratables** e **intratables**:

- Un problema tratable se puede resolver en un tiempo polinomial o razonable.

- Un problema intratable es aquel que no puede ser resuelto en un tiempo polinomial.

Un problema **indecidible** es aquel que no se puede resolver mediante un algoritmo, sino solamente obtener soluciones aproximadas que pueden ser o no correctas.

### Clases de complejidad

**Clasificación P** se compone de los problemas tratables en los que existe un algoritmo que los puede resolver eficientemente, cuya función de complejidad está acotada por algún polinomio.

**Clasificación NP** es el conjunto de problemas cuya solución se puede **verificar** en tiempo polinomial, pero no necesariamente se pueden resolver en un tiempo razonable.

Habiendo entendido ambas partes, podemos comprender la conjetura $P≠NP$.

Sabemos que $P$ son aquellos problemas que se pueden resolver en un tiempo razonable, en cambio los problemas $NP$ son aquellos que se resuelven en un tiempo no determinístico, por lo cual no se sabe cuánto tiempo puede llevar encontrar dicha solución. En consecuencia, no es posible afirmar que $P = NP$, ya que no es seguro que todos los problemas de NP puedan resolverse en tiempo polinomial. Esto no ha sido demostrado del todo, sin embargo, es poco probable que esto sea cierto, por lo cual se cree que $P≠NP$.








## ¿Cómo se prueba que un problema es NP-Hard usando reducciones?

Si es posible encontrar un algoritmo que transforme un problema $A$ NP-Hard en un problema $B$ en tiempo polinomial, decimos que el problema $A$ se **reduce** al problema $B$.

Para demostrar que un problema $B$ es **NP-hard** se aplica reducción:

1. Selecionamos un problema $A$ que sabemos que es **NP-hard**.

2. Si el problema $A$ es reducible a $B$, entonces el problema $B$ es **NP-hard**.

En consecuencia, llegamos a la conclusion de que $B$ es un problema **NP-hard**.

## Reducción al problema del conjunto independiente máximo

### Problema del conjunto independiente máximo

**Entrada**: Un grafo no dirigido $G=(V,E)$.

**Salida**: Conjunto máximo de vértices tal que ninguno de los vértices es adyacente a otro.

Este problema consiste en determinar el conjunto máximo que se puede formar con los vértices de un grafo tal que los nodos del conjunto no sean adyacentes entre sí. Por ejemplo, los vértices azules en la siguiente figura forman un conjunto independiente máximo:



Sabiendo que este problema es NP-hard, podemos aplicar reducción para demostrar que el problema de la cobertura de vértices también lo es.

### Reducción

Un primer acercamiento a esta demostración, es comprender que ambos problemas parecen ser opuestos. Por un lado, resolver el problema de la cobertura de vértices implica **minimizar** la cantidad de vértices para conectar todos los arcos del grafo. Por otra parte, el problema del conjunto independiente busca **maximizar** la cantidad de vértices no adyacentes, evitando las conexiones entre ellos.

#### **Teorema 1**

*Sea $G$ un grafo no dirigido, $n$ la cantidad de vértices de $G$, $V^{\prime}$ una cobertura de vértices de $G$ y $S$ un conjunto independiente de $G$. Existe un conjunto $V^{\prime}$ con $k$ vértices si y solo si existe un conjunto $S$ con $n-k$ vértices.*

Supongamos que existe una cobertura de vértices $V^{\prime}$ con $k$ vértices. Entonces, según el teorema, debe existir un conjunto independiente con $n-k$ vértices. Llamaremos $V \backslash V^{\prime}$ al conjunto de vértices del grafo que no pertenecen a la cobertura de vértices. Consideremos cualquier par de vértices $u, v \in V \backslash V^{\prime}$. Podemos afirmar que no existe ningún arco entre los nodos $u$ y $v$. Por contradicción, si existiera un arco $(u, v)$, este debería ser cubierto por $V^{\prime}$, y en consecuencia, el par de nodos $u, v$ también debería ser parte del conjunto. Por definición, sabemos que $u, v \notin V^{\prime}$, por lo tanto, este caso no puede ocurrir.

Luego, como no existen arcos entre los vértices del conjunto $V \backslash V^{\prime}$, sabemos que los vértices que no están en la cobertura formarán un conjunto independiente $S$. Por definición, dado que $V \backslash V^{\prime}$ corresponde a la resta del conjunto $V$ con $n$ vértices menos el conjunto $V^{\prime}$ con $k$ vértices, se cumple que el tamaño del conjunto $S$ será de $n-k$.

Dado que el teorema supone una doble implicación, también debemos demostrar que, si existe un conjunto independiente $S$ con $n-k$ vértices, existirá una cobertura de vértices de tamaño $k$. Sea $E$ el conjunto de los arcos del grafo $G$ y $V \backslash S$ el conjunto de vértices que no pertenecen al conjunto independiente. Para cualquier arco $(u, v)$, se cumple que al menos uno de los vértices no forma parte de $S$, dado que no pueden haber nodos adyacentes en este conjunto. Por lo tanto, uno de estos vértices será parte del conjunto $V \backslash S$. Luego, como cada arco del grafo se conecta con al menos uno de estos vértices, se cumple que $V \backslash S$ es una cobertura de vértices de tamaño $k$. $\Box$

#### **Teorema 2**

*El problema del conjunto independiente máximo es reducible en tiempo polinomial al problema de la cobertura de vértices.*

Probaremos el teorema demostrando que es posible resolver el problema del conjunto independiente máximo llamando a una subrutina que resuelva el problema de la cobertura de vértices. 

Dado un grafo $G = (V, E)$ como entrada al problema del conjunto independiente máximo, podemos llamar directamente a un algoritmo que encuentre la menor cobertura de vértices de $G$, sin aplicar ningún tipo de preprocesamiento especial. Esta subrutina retornará un conjunto $V^{\prime}$ de tamaño $k$. En base al Teorema 1, sabemos que el conjunto complementario de $V^{\prime}$ es un conjunto independiente $S$ de tamaño $n-k$. Por lo tanto, basta con obtener los vértices de $G$ que no pertenezcan al conjunto $V^{\prime}$ para encontrar los vértices del conjunto independiente $S$. Esto se puede realizar fácilmente en tiempo polinomial, recorriendo los vértices del grafo y descartando aquellos que se encuentren en $V^{\prime}$.

Por último, observemos que la subrutina encontrará una cobertura de vértices **mínima**. Luego, se cumple que el complemento de este conjunto, es decir, el conjunto independiente $S$, será **máximo**, encontrando así una solución correcta al problema. $\Box$

### Algoritmos para resolver el problema

#### Algoritmo de aproximación en tiempo polinomial

El siguiente algoritmo encuentra una solución **aproximada** al problema de la cobertura de vértices:

1. Inicializamos una lista vacía $R$ para guardar los vértices que forman la cobertura.

2. Añadimos todos los arcos del grafo a una lista $E$.

3. Mientras $E$ no esté vacía, elegimos un arco $(u, v)$ y eliminamos todos los arcos que incidan sobre $u$ o $v$, incluyendo el mismo arco $(u, v)$. Añadimos los nodos $u$ y $v$ a la lista $R$.

4. Retornamos la lista $R$.

Iterar por los arcos del grafo tiene un tiempo de ejecución $O(E)$, mientras que añadir los vértices a la lista $R$ tiene un tiempo de ejecución $O(V)$. Por lo tanto, el algoritmo anterior presenta un tiempo de ejecución total de $O(V+E)$.

Se puede probar que esta solución retorna una cobertura de vértices cuyo tamaño nunca es mayor al doble del tamaño de la mínima cobertura de vértices. Por lo tanto, si se quiere encontrar una solución en un tiempo razonable, puede ser una buena opción ocupar este algoritmo.

#### Algoritmo exacto en tiempo exponencial

Podemos encontrar la menor cobertura de vértices de manera exacta utilizando un algoritmo de fuerza bruta. La idea de este algoritmo consiste en probar cada combinación posible de vértices, comenzando por aquellas combinaciones con una menor cantidad de nodos hasta encontrar un conjunto que resuelva el problema.

1. Verificamos si una combinación de $m$ vértices corresponde a una cobertura de vértices, comenzando desde $m = 1$.

2. De no encontrar una solución, aumentamos el valor de $m$ en 1 y repetimos el paso anterior.

3. Una vez encontrada la solución, dejamos de iterar y retornamos la combinación de vértices.

El tiempo de ejecución de este algoritmo es $O(2^V \cdot E)$, dado que se procesa cada una de las $2^V$ combinaciones posibles de vértices y se verifica que corresponda a una solución pasando por todos los arcos del grafo. 

# Problema: Subset-Sum

**Entrada**: Conjunto $S$ de números enteros y $T$ como valor o suma objetivo.

**Salida**: Subconjunto $S^{\prime} \subseteq S$ tal que la suma de sus elementos corresponda al valor de $T$

Se muestra el siguiente ejemplo para entender el funcionamiento de la solución del problema.

Sea $S= \{−7,−3,−2,5,8\}$ el conjunto de números enteros y $T = 0$, se verifica que el problema tiene una solución factible ya que $\{−3,−2,5\}$ es un subconjunto $S^{\prime} \subseteq S$ tal que la suma de sus elementos es 0 correspondiente al valor $T$ buscado.

A simple vista, la solución del problema puede verse bastante sencilla pero la complejidad en tiempo de ejecución de subset-sum depende de dos parámetros: $n$ el número de enteros de entrada y $L$ la precisión del problema, expresada como el número de valores posicionales binarios que se necesitan para plantear el problema. Si estos dos parametros son del mismo orden el problema se torna más difícil y por ende de complejidad exponencial.

Dado este problema, se buscará reducirlo al problema de la mochila para demostrar que este último también es NP-hard.

## Reducción al problema de la mochila

### Problema: Knapsack

Se define una mochila con una capacidad $m$ de $n$ unidades de peso. Además de una secuencia de $n$ elementos, cada uno con un peso $p$ y un valor $v$. Se busca una subsecuencia de elementos de forma que se maximice la ganancia y no se sobrepase el límite de unidades de peso.

**Entrada**: Capacidad de la mochila $m$ y una secuencia de elementos $E$.

**Salida**: Valor máximo que se puede obtener en la mochila sin superar el límite de peso $n$.

### Reducción 

Para el knapsack se requiere maximizar la ganancia obtenida con un límite de peso de la mochila y un valor establecido para cada objeto por lo que el problema depende de dos variables.

Se considera que para cada objeto, el valor y el peso son igual a cada número del conjunto de entrada de *subset-sum*.

Entonces, si se tiene un conjunto $S$ de números enteros (subset-sum), este mismo conjunto $S$ correponderá a una secuencia de $n$ objetos (knapsack) donde cada elemento del conjunto tendrá un peso $p$ y un valor $v$ igual al valor de cada número de $S$, donde $T$ es el valor buscado en la suma de los elementos del subconjunto $S^{\prime}$ (subset-sum) que será igual a la capacidad máxima $m$ de la mochila.

Al resolver el problema de la mochila se buscará maximizar la ganancia obtenida con los objetos o datos disponibles del conjunto $n$. Dado que $p_{i}=v_{i}=S_{i}$, habremos encontrado a su vez el subconjunto $S^{\prime} \subseteq S$ el cual contendrá la suma máxima $h \leq T$. Si $h = T$, entonces $S$ corresponde a una solución para el problema de la suma de subconjuntos.  



### Algoritmos para resolver el problema

#### Algoritmo de aproximación en tiempo polinomial

El siguiente algoritmo encuentra una solución aproximada al problema Subset-Sum:

1. Hacer un ciclo donde $i$ va de $1$ a $n+1$ (donde $n$ es la cantidad de elementos del conjunto $S$).
2. En cada iteración $i$, se realiza una combinatoria de $n$ sobre $i$ donde se obtendrán todas las combinaciones de números posibles de tamaño $i$ comprobando si la suma de alguna de ellas corresponde al valor buscado $T$.
3. De encontrarse se retorna el subconjunto $S^{\prime}$ que cumpla con el valor de $T$.

Hay varias maneras de resolver la suma de subconjuntos en tiempo exponencial sobre $n$. El algoritmo más simplista verificaría todos los posibles subconjuntos de $n$ y, para cada uno de ellos, compararía la suma al total buscado. El tiempo de ejecución es de orden $O(2^n \cdot n)$, dado que hay $2n$ subconjuntos y, para verificar cada subconjunto, tenemos que sumar $n$ elementos.
